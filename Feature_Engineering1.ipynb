14.What is an Optimizer? What are different types of optimizers? Explain each with an example.
-  In machine learning, an optimizer is an algorithm that iteratively adjusts the model‚Äôs parameters (e.g. weights) in order to minimize the loss (error) function. It computes and applies updates based on gradients of the loss until the model converges to an acceptable solution.

The diffrent types of optimizers are given as below

1. Gradient Descent (Batch GD)
Definition: Computes the gradient of the loss across the entire dataset and updates parameters accordingly

Example: Used in simple linear regression problems where the dataset is small enough to compute full gradients efficiently.

2. Stochastic Gradient Descent (SGD)
Definition: Updates parameters using the gradient from a single random sample (or mini-batch).

Pros: Faster convergence per iteration; well-suited for large datasets.

Cons: Noisy updates can cause oscillation; requires careful learning rate tuning
Example: Widely used to train deep networks like CNNs on large datasets such as ImageNet.

3. SGD with Momentum (including Nesterov Accelerated Gradient)
Definition: Introduces a velocity term that accumulates past gradients to smooth updates and accelerate convergence.

Pros: Reduces oscillations, speeds up convergence in ravines.

Cons: Adds hyperparameter (momentum factor); may overshoot if tuned poorly
Example: Useful in training RNNs or deep architectures where loss landscapes can be noisy and curved.

4. Adagrad
Definition: Adapts learning rate per parameter based on historical gradients‚Äîlarge updates for rare features, smaller for frequent ones.

Pros: No need for manual per-feature tuning; good for sparse data.

Cons: Learning rate decays too quickly, potentially halting learning prematurely
Example: Effective in NLP tasks like training word embeddings where certain words appear sparsely.
